# Computation of a Shapley value  

It is simple to conduct the contribution of features within a Linear Regression model versus the prediction. A Multiple Linear Regression model can be written such as:

$f(\hat{x}) = \beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_kX_k + e $ (Eqn 1)

+ Where ${f}(\hat{x})$ is modeled as a linear function of $k$ explanatory variables, $ X_1, X_2, \ldots, X_k$.   
+ $\beta_k $ is the weight corresponding to feature $k$.   

Thus the contribution $\phi_m$ of the m-th feature on the prediction can be written as:

$\phi_m{(\hat{f})}$  = $\beta_mx_m - E(\beta_mX_m) = \beta_mx_m - \beta_mE(X_m)$   (Eqn 2) 


+ Where $E(\beta_mX_m)$ is the average effect estimate for feature $m$.   
+ In order to work out the contribution of each feature, the contribution is the difference between the **feature effect** minus the **average effect**.   
+ Then we will sum all the feature contributions, which would equate to the following formula shown in Eqn 3:

Sum of all the feature contributions = $f(\hat{x}) - E(f(\hat{X}))$  (Eqn 3)

However, this is difficult for more complex models, which we will be using in this project, such as XGBoost.

The Shapley Value can be defined via a value function ${val}$ of players in $S$. The Shapley Value of a feature is its contribution to the payout, weighted and summed over all possible feature value combinations:

$\phi_m$(val)  = $\sum_{S\subseteq{(1,...,p)}\backslash {(m)}}$     $\frac{|S|!(p-|S|-1)!}{p!}$   $(\text{val}(S\cup\{m\})$   $-$    $\text{val(S))}$ (Eqn 4)



Where:

+ $m$ is the m -th feature
+ $S$ is the subset of features used
+ $p$ is the number of features used
+ $\text{val}(S\cup\{m\})$ is the combination of the subset of features combined with the m-th feature
+ $\text{val}(S)$  is the combination of the subset of features without the m-th feature

Therefore, $\text{val}(S\cup\{m\}) - \text{val}(S))$ is the difference between including the m-th feature versus not including it. These values are then weighted and summed over all possible feature value combinations. Hence, equation (4) shows how to compute a Shapley Value. [[1]](https://christophm.github.io/interpretable-ml-book/shapley.html)

# What are Shapley Values? (A really speedy intro) 

We can analyse Shapley Values to understand how  **<span style="color: red;">each feature</span>** is contributing towards the **<span style="color: red;">prediction</span>**. 

In this example I am utilising the **California housing dataset** where we will be predicting the **House Value** from a range of features. Once we have built our model we want to understand **how** each of the **features** contribute towards the **prediction**. 

When you are building a model within industry, you can probably imagine you will collect many features, so to avoid computational complexity, SHAP is a really good way to reduce the number of features in the model.   

Shapley values are a really good way to **understand what the features are doing in the model** and **how they are influencing the prediction**.

Shapley Values are an established idea from Game Theory, whereby the prediction can be explained by assuming that each feature value of the data is a player in a game where the prediction is the payout. 

Shapley Values show how to distribute the payout among the features within the model [[1]](https://christophm.github.io/interpretable-ml-book/shapley.html). 





**Example:** Imagine we only had **four features** in our model to **predict House Value**, and for a certain house in the dataset had the following criteria; 
+ House is 8 years old.    
+ Average number of rooms is 3. 
+ Average number of bedrooms is 1.  
+ Population is 2000.  

Imagine out of all houses in the data the **predicted average House Value is  £175K**, however the feature values shown worked together to produce a prediction of £150K.   

Our goal is to explain the difference between the model's prediction of £150K for this example, and the average prediction of £175K (a difference of £25K).   

Shapley Values look at the **average marginal contribution** of a feature **across all possible combinations**.   
It works in the following way:  
1. Take one particular feature.  
2. Look at all combinations of other features with that particular feature.  
3. Compute the House Value with and without that particular feature.  
4. The Shapley Value is the (weighted) average of marginal contributions.  



### In summary

- They are used to **understand what features should be put into a model**. An alternative to other feature importance methods (e.g. LIME, Permutation importance, random forest feature importance etc.).   


- They can be used to understand **how** each feature is contributing towards the dependent variable/Target/Outcome variable.  


- An advantage of shapley values is you are **able to understand what is happening within a black box ML model**, as well as understanding which are the best features to input into the model.  


- SHAP makes transparent the correlations picked up by predictive ML models, but making correlations transparent does not make them causal. Here is an article on this [here](https://towardsdatascience.com/be-careful-when-interpreting-predictive-models-in-search-of-causal-insights-e68626e664b6)

# Build Train, Test sets 

```python
X_columns = ['HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude','MedInc']
y_columns = ['MedHouseVal']
```

```python
X = california_housing.frame[X_columns]
X.head()
```

```python
y = california_housing.frame[y_columns]
y.head()
```

```python
# Select train, test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 12)
```

# Build a Model

First of all, you need to build a ML model. 

Here I am building a simple **XGboost** machine learning model (but you could build something very simple like a Linear Regression).  

```python
model = xgboost.XGBRegressor(random_state=12, verbosity=0).fit(X_train, y_train)
```

# Let's look at some beautiful SHAP charts 😻

The **Explainer** uses Shapley values to explain a tree based machine learning model.
There is other Explainers such as DeepExplainer for Deeplearning.

```python
explainer = shap.Explainer(model)
shap_values_train = explainer(X_train) #This computes all the shapley values for the training set
```

Let's look what the explainer produces

```python
explainer(X_train) 
```

The shap_values_train shows three arrays.

**.values**

These are the SHAP values (The average of all marginal contributions worked out when you include and exclude the feature for all combinations).  

**.base_values**

Base values is the average model prediction (based on provided training data).  

**.data...**

This is the input data for the model (the features).  We have 8 features in this model, so per house, there will be 8 data points.


.base_value is the average model prediction over the training set (Can be computed in the following way too - it is very slightly different due to rounding)

```python
model.fit(X_train, y_train)
model.predict(X_train).mean() 
```

# 1. Global feature importance bar plots (Global = for all houses combined in the training dataset)

Passing a matrix of SHAP values to the bar plot function creates a global feature importance plot, where the global importance of each feature is taken to be the **mean absolute SHAP value** for that feature over all the given samples. [[2]](https://shap.readthedocs.io/en/latest/example_notebooks/api_examples/plots/bar.html)



- X-axis shows the mean absolute of the SHAP values. (Hence, doesn't include whether the feature is positively or negatively effecting the prediction).   


- Y-axis shows the features, ordered by the mean absolute of the SHAP values.


- By default is shows only 10 bars, which you can change.  


```python
shap.plots.bar(shap_values_train)
```

The bar plot shows that **Average Rooms** has the highest mean absolute SHAP value i.e. the highest effect on the prediction.  

You can add in max_display parameter field.

```python
shap.plots.bar(shap_values_train, max_display=5)
```

Or you can have the plot without the labels

```python
shap.summary_plot(shap_values_train, X_train, plot_type="bar")
```

# 2. Local feature importance plots (one house at a time)

- Passing just one row of SHAP values to the bar plot function creates a local feature importance plot, where the bars are the SHAP values for each feature.  


- X-axis is the SHAP value for the house from row 300.


- Y-axis are the feautures ordered in descending order based on the absolute SHAP value. 

```python
shap.plots.bar(shap_values_train[300], max_display=10)
```

Some insights from this chart:  
+ The feature **Latitude** is pushing up the prediction House Value.  
+ The feature **Median Income** is pushing up the prediction House Value.  
+ The feature **Longitude** is pushing down the prediction House Value.  
    

Let's double check the theory again...

The sum of all the shap values for row 300

```python
sum_of_shap_300 = shap_values_train.values[300].sum()
print(sum_of_shap_300)
```

The average prediction out of all values in the training data (SHAP base value)

```python
shap_values_train_base_value = shap_values_train.base_values[0]
print(shap_values_train_base_value)
```

Shap base value + SHAP values = Model prediction

```python
shap_values_train_base_value + sum_of_shap_300
```

Model prediction for row 300

```python
model_prediction_300 = model.predict(X_train)[300]
print(model_prediction_300)
```

Local feature importance can also be explored via waterfalls (same chart as above but with the feature values + a couple of extra bits of information i.e. f(x), E[f(x)] ).   

```python
shap.plots.waterfall(shap_values_train[300],max_display=3)
```

f(x) = the model prediction for row 300 (2.216 i.e. 221,600)  
E[f(x)] = the base value i.e. the average model prediction over the training set (2.063 i.e. 206,300)  



Local feature importance can also be explored via force plots

```python
shap.initjs()
shap.plots.force(shap_values_train[300])
```

IMO the above plot is confusing to interpret vs. the others I have shown.  

# 3. Summary Plots &#x1F60D;



- Each point on each row represent a house in the train set.   


- The **X-axis** demonstrates the range of SHAP values (the impact on the model output).


- The **Y-axis** shows the feature names, ordered by the mean absolute SHAP values, highest being at the top.   


- The colour represents the values of the raw data of the features, low values of that feature equal **<span style="color: blue;">blue</span>**, high values of that feature = **<span style="color: red;">red</span>**. 

```python
shap.summary_plot(shap_values_train, features=X_train, feature_names=X_train.columns,max_display=14)
```

Positive SHAP values push the **prediction up.**   
Negative SHAP values push the **prediction down.**  

I can talk you through some key insights we get from this chart.

- The top feature is Latitude. It shows that lower values of this feature predict a higher House Value, and higher values of this feature predict a lower House Value (so if your house is south calfornia, the house value goes up).


- The second top feature is Longtitude - Lower values of Longtitude produce higher House Value. Higher values of Longtitude produce lower values of House Value.  

- The third top feature is Median Income. Higher values of this push the prediction up, and lower values of Income push the prediction down.  


# 4. Interaction Plots

A dependence plot is a scatter plot that shows the effect a single feature has on preditions made by the model. 

- Each dot is a single house from the training dataset.   


- The X-Axis shows the value of the feature i.e. from the feature data we provided to train the model. (It goes between 0 rooms and 140 rooms).      


- The Y-Axis shows the SHAP values for that feature, which represents how much knowing that feature's value changes the output of the model for that sample's prediction.    


- The colour corresponds to the second feature that may have an interaction effect with the feature we have plotted. If there is an interaction effect, it will show a distinct vertical pattern of colouring.



I couldn't find any interactions within these features, but wanted to show the plots...

```python
shap.dependence_plot('AveRooms', # Feature of interest
                     shap_values_train.values, # Shap values
                     X_train, # Training data
                     interaction_index="HouseAge")
```

There is some better examples [here.](https://shap-lrjball.readthedocs.io/en/latest/example_notebooks/plots/dependence_plot.html)

# Appendix

What is Random Forest vs. XGBoost?

What is a XGBoost model?   
XGBoost utilises boosting, Boosting is the technique in which each model tries to improve on the error from the previous model. Boosting tries to reduce bias.  Gradient Boosting is a specific technique of boosting in which gradient descent optimization is used to minimise the loss function while creating trees

Whereas Random forest is a technique that uses Bagging. Bagging is the technique that builds different models in parallel using random subsets of data and deterministically aggregates the predictions of all predictors. Bagging tries to tackle the over-fitting problem.
